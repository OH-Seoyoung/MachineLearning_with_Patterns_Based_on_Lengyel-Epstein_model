```python
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image 
import random
```

```python
# Make dataset
x_orig = []
y_orig = np.zeros((1,120))
for i in range(0, 36):
    for j in range(i*120 + 1, i*120 + 121) :
        img = Image.open('dataset/{0}/pattern_{1}.jpg'.format(i, j)) 
        data = np.array(img)
        x_orig.append(data)
        
for i in range(1,36):
    y_orig = np.append(y_orig, np.full((1, 120),i), axis = 1)     
x_orig = np.array(x_orig)
y_orig = y_orig.T
```

```python
print(x_orig.shape)
print(y_orig.shape)
```

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    image = np.reshape(x_orig[i+4200,:] , [64,64])
    plt.imshow(image)
    plt.xlabel(y_orig[i+4200])
plt.show()
```

```python
# Flatten the training and test images
x_flatten = x_orig.reshape(x_orig.shape[0], -1)

# Normalize image vectors
X = (2/255) * x_flatten - 1

# Explore dataset 
print ("number of examples = " + str(X.shape[0]))
print ("X shape: " + str(X.shape))
print ("Y shape: " + str(y_orig.shape))
```

```python
def gradient_vec(X):
    g_X_r = np.gradient(X, axis = 1)
    g_X_c = np.gradient(X, axis = 0)
    g_X = g_X_r**2 + g_X_c**2
    return g_X
```

```python
X1 = np.var(X - X**3, axis=1)
X2 = np.var(1-X**2, axis = 1)
```

```python
X1 = X1.reshape((4320,1))
X2 = X2.reshape((4320,1))

print(X1.shape)
print(X2.shape)
```

```python
plt.rcParams["figure.figsize"] = (25,14)
```

```python
plt.scatter(X1, X2)
plt.axis([min(X1), max(X1), min(X2), max(X2)])
plt.xlabel("Var of X-X^3", fontsize = 25)
plt.ylabel("Var of 1-X^2", fontsize = 25)
plt.savefig('graph4/all.jpg', dpi=300)
plt.show()
```
